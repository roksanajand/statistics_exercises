### Zadanie 1: Estymacja wartoci oczekiwanej [8 pkt]

Przygotuj funkcj `mean_estimation()`, kt贸ra bdzie dokonywaa estymacji przedziaowej wartoci oczekiwanej dla danych wejciowych w postaci szeregu szczeg贸owego zgodnie ze schematem z zaczonego do notatnika zestawu wzor贸w.

Opr贸cz zwracanych wartoci granic przedziau funkcja powinna wywietla nastpujcy komunikat:

` nale偶y do przedziau [X, Y] przy zao偶eniu poziomu ufnoci 1- = Z`

gdzie X, Y i Z s automatycznie uzupeniane przez funkcj, a X i Y dodatkowo sformatowane w taki spos贸b, 偶eby wywietlay si z dokadnoci do 4 miejsc po przecinku.

```python
def mean_estimation(data, alpha, population_std="unknown"):
    srednia = data.mean().values[0] #rednia , same wartosci i 0 okresla pierwszy element
    n = len(data)  #rozmiar pr贸bki
    
    if population_std != "unknown":
        #znane std
        i = sp.stats.norm.ppf(1-alpha/2)
        margines = i*population_std/np.sqrt(n)
    else:
        #nieznane std
        s = data.std().values[0]  #standardowe std
        t = sp.stats.t.ppf(1-alpha/2, n-1)  #wartosc krytyczna
        margines = t*s/np.sqrt(n)           #korzystamy z std pandas wic ddof=1, nie musimy tutaj odejmowac

    #okreslanie granic
    X = srednia-margines #dolna
    Y = srednia+margines #gorna
    Z = 1-alpha          #poziom ufnosci

    #Drukowanie i wywoanie
    print(f"渭 nale偶y do przedziau [{X:.4f}, {Y:.4f}] przy zao偶eniu poziomu ufnoci 1-伪 = {Z}")
    return(X,Y)
```

```python
### Kom贸rka testowa

# Test na test_data_1 przy zao偶eniu, 偶e znane jest odchylenie standardowe populacji, z kt贸rej zostaa pobrana pr贸bka losowa
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.1, population_std=0.2), (4.905619719495418, 5.05273990041144))))
print()
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.05, population_std=0.2), (4.891527555895771, 5.066832064011087))))
print()
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.02, population_std=0.2), (4.875142370240081, 5.083217249666777))))
print()

# # Test na test_data_1 przy zao偶eniu, 偶e nie jest znane odchylenie standardowe populacji, z kt贸rej zostaa pobrana pr贸bka losowa
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.1), (4.913356369231727, 5.045003250675131))))
print()
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.05), (4.899504007229391, 5.058855612677467))))
print()
assert(np.all(np.isclose(mean_estimation(test_data_1, 0.02), (4.882508507351398, 5.07585111255546))))
```

### Zadanie 2: Estymacja wariancji [8 pkt]

Przygotuj funkcj `variance_estimation()`, kt贸ra bdzie dokonywaa estymacji przedziaowej wariancji dla danych wejciowych w postaci szeregu szczeg贸owego zgodnie ze schematem z zaczonego do notatnika zestawu wzor贸w.

Opr贸cz zwracanych wartoci granic przedziau funkcja powinna wywietla nastpujcy komunikat:

`^2  nale偶y do przedziau [X, Y] przy zao偶eniu poziomu ufnoci 1- = Z`

gdzie X, Y i Z s automatycznie uzupeniane przez funkcj, a X i Y dodatkowo sformatowane w taki spos贸b, 偶eby wywietlay si z dokadnoci do 4 miejsc po przecinku.

```python
def variance_estimation(data, alpha, population_mean="unknown"):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i pojedyncz kolumn "X" z wartociami pr贸by losowej,
    alpha: float
        Warto okrelajca poziom istotnoci.
    population_mean: str or float
        Je偶eli warto oczekiwana populacji nie jest znane to parametr przyjmuje warto "unknown",
        w przeciwnym wypadku jest to warto okrelajca warto oczekiwan populacji.
    
    Returns
    -------
    var_low: float
        Dolna granica wyliczonego przedziau ufnoci.
    var_high: float
        G贸rna granica wyliczonego przedziau ufnoci.
    """
    sample_data=data['X'].values
    n=len(sample_data) #rozmiar 

    if population_mean == "unknown":
        #nieznane population_mean, licze ze wzoru x
        x=sample_data.mean()
        s_squared=(1/n)*np.sum((sample_data-x)**2)
    else:
        #znane population_mean
        s_squared=(1/n)*np.sum((sample_data-float(population_mean))**2)

    #granice
    var_low=(n*s_squared)/sp.stats.chi2.ppf(1-alpha/2, n-1)
    var_high=(n*s_squared)/sp.stats.chi2.ppf(alpha/2, n-1)
    Z = 1 - alpha #poziom ufnoci

    print(f"^2  nale偶y do przedziau [{var_low:.4f}, {var_high:.4f}] przy zao偶eniu poziomu ufnoci  1-伪 = {Z:.4f}")

    return var_low,var_high
```

```python
### Kom贸rka testowa

# Test na test_data_1 przy zao偶eniu, 偶e znane jest odchylenie standardowe populacji, z kt贸rej zostaa pobrana pr贸bka losowa
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.1, population_mean=5), (0.01855573624556403, 0.05528660848892888))));
print()
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.05, population_mean=5), (0.017025745016269588, 0.06280068548350207))));
print()
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.02, population_mean=5), (0.015455150810030724, 0.07328116757245808))));
print()

# Test na test_data_1 przy zao偶eniu, 偶e nie jest znane odchylenie standardowe populacji, z kt贸rej zostaa pobrana pr贸bka losowa
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.1), (0.018268125369359626, 0.05442967510189219))));
print()
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.05), (0.0167618487538223, 0.06182728513230527))));
print()
assert(np.all(np.isclose(variance_estimation(test_data_1, 0.02), (0.015215598500840796, 0.07214532146342346))));
```

### Zadanie 3: Estymacja wartoci wsp贸czynnika korelacji liniowej [6 pkt]

Przygotuj funkcj `correlation_estimation()`, kt贸ra bdzie dokonywaa estymacji przedziaowej wartoci wsp贸czynnika korelacji liniowej Pearsona dla danych wejciowych w postaci szeregu szczeg贸owego zgodnie ze schematem z zaczonego do notatnika zestawu wzor贸w.

Opr贸cz zwracanych wartoci granic przedziau funkcja powinna wywietla nastpujcy komunikat:

`r nale偶y do przedziau [X, Y] przy zao偶eniu poziomu ufnoci 1- = Z`

gdzie X, Y i Z s automatycznie uzupeniane przez funkcj, a X i Y dodatkowo sformatowane w taki spos贸b, 偶eby wywietlay si z dokadnoci do 4 miejsc po przecinku.

```python
def correlation_estimation(data, alpha):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i dwie kolumny "X" i "Y" z wynikami pr贸by losowej.
    alpha: float
        Warto okrelajca poziom istotnoci.
    Returns
    -------
    r_corr_low: float
        Dolna granica wyliczonego przedziau ufnoci.
    r_corr_high: float
        G贸rna granica wyliczonego przedziau ufnoci. 
    """   

    n = len(data)  # liczba obserwacji

    # Obliczenie rednich dla X i Y
    X_mean = data['X'].mean()
    Y_mean = data['Y'].mean()

    # Obliczenie estymatora wsp贸czynnika korelacji Pearsona
    licznik = np.sum((data['X'] - X_mean) * (data['Y'] - Y_mean))
    mianownik = np.sqrt(np.sum((data['X'] - X_mean)**2) * np.sum((data['Y'] - Y_mean)**2))
    r = licznik / mianownik
    
    # Obliczenie uwarunkowania
    u = sp.stats.norm.ppf(1 - alpha / 2)

    # Obliczenie granic przedziau ufnoci
    r_corr_low = r - u * (1 - r**2) / np.sqrt(n)
    r_corr_high = r + u * (1 - r**2) / np.sqrt(n)

    
    # Formatowanie wartoci X i Y
    X = round(r_corr_low, 4)
    Y = round(r_corr_high, 4)
    Z=1-alpha

    print(f"r nale偶y do przedziau [{X}, {Y}] przy zao偶eniu poziomu ufnoci 1-{alpha} = {Z}")

    return r_corr_low, r_corr_high
```

```python
### Kom贸rka testowa
assert np.all(np.isclose(correlation_estimation(test_data_2, 0.05), (0.9494696249316307, 0.9692185148958979)))
assert np.all(np.isclose(correlation_estimation(test_data_2, 0.02), (0.9476237555239514, 0.9710643843035772)))
```

### Zadanie 4: Estymacja wsp贸czynnik贸w r贸wnania regresji liniowej [8 pkt]

Przygotuj funkcj `interval_linear_regression_coefficients_estimation()`, kt贸ra bdzie dokonywaa estymacji przedziaowej wartoci wsp贸czynnik贸w r贸wnania regresji liniowej dla danych wejciowych w postaci szeregu szczeg贸owego zgodnie ze schematem z zaczonego do notatnika zestawu wzor贸w.

Opr贸cz zwracanych wartoci granic przedziau funkcja powinna wywietla nastpujcy komunikat:

`a nale偶y do przedziau [X1, Y1] przy zao偶eniu poziomu ufnoci 1- = Z1` </br>
`b nale偶y do przedziau [X2, Y2] przy zao偶eniu poziomu ufnoci 1- = Z2`

gdzie X1, X2, Y1, Y2, Z1 i Z2 s automatycznie uzupeniane przez funkcj, a X1, X2, Y1 i Y2 dodatkowo sformatowane w taki spos贸b, 偶eby wywietlay si z dokadnoci do 4 miejsc po przecinku.

```python
def linear_regression_coefficients_estimation(data, alpha):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i dwie kolumny "X" i "Y" z wynikami pr贸by losowej.
    alpha: float
        Warto okrelajca poziom istotnoci.
    Returns
    -------
    a_ci: tuple
        Zmienna typu tuple zawierajaca granice przedziau ufnoci parametru a (a_low, a_high).
    b_ci: tuple
        Zmienna typu tuple zawierajaca granice przedziau ufnoci parametru b (b_low, b_high).
    """   
    # YOUR CODE HERE
    n = len(data)  # liczba obserwacji

    # Obliczenie rednich dla X i Y
    X_mean = data['X'].mean()
    Y_mean = data['Y'].mean()

    # Obliczenie estymatora wsp贸czynnika a
    licznik = np.sum((data['X'] - X_mean) * (data['Y'] - Y_mean))
    mianownik = np.sum((data['X'] - X_mean)**2)
    a = licznik / mianownik

    # Obliczenie estymatora wsp贸czynnika b
    b = Y_mean - a * X_mean

    # Obliczenie niepewnoci dla a
    su = np.sqrt(1 / (n - 2) * np.sum((data['Y'] - (a * data['X'] + b))**2))
    Da = su / np.sqrt(np.sum((data['X'] - X_mean)**2))

    # Obliczenie niepewnoci dla b
    Db = su * np.sqrt(np.sum((data['X'])**2) / (n * np.sum((data['X'] - X_mean)**2)))

    # Obliczenie wartoci krytycznych dla przedzia贸w ufnoci
    t_statistic = sp.stats.t.ppf(1 - alpha / 2, n-2)

    # Obliczenie granic przedzia贸w ufnoci dla a
    a_low = a - t_statistic * Da
    a_high = a + t_statistic * Da

    # Obliczenie granic przedzia贸w ufnoci dla b
    b_low = b - t_statistic * Db
    b_high = b + t_statistic * Db

    # Formatowanie wartoci
    X1, X2, Y1, Y2, Z1, Z2 = round(a_low, 4), round(a_high, 4), round(b_low, 4), round(b_high, 4), round(1-alpha, 4), round(1-alpha, 4)

    print(f"a nale偶y do przedziau [{X1}, {X2}] przy zao偶eniu poziomu ufnoci 1-{alpha} = {Z1}")
    print(f"b nale偶y do przedziau [{Y1}, {Y2}] przy zao偶eniu poziomu ufnoci 1-{alpha} = {Z2}")

    return (a_low, a_high), (b_low, b_high)
```

```python
### Kom贸rka testowa
assert np.all(np.isclose(linear_regression_coefficients_estimation(test_data_2, 0.1), ((1.90330932143819, 2.024457061652292), (-5.331039505544365, -4.7958291264947706))))
print()
assert np.all(np.isclose(linear_regression_coefficients_estimation(test_data_2, 0.05), ((1.8916219551002786, 2.0361444279902035), (-5.382672328053067, -4.744196303986069))))
```
